```python
# SECTION 1 — Imports
import torch
from torchvision.models import resnet50
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
# SECTION 2 — Load Black-Box Model
model = resnet50()
model.load_state_dict(torch.load("model/blackbox_model.pt"))
model.eval()

print("Model loaded successfully")
```

```python
# SECTION 3 — Preprocessing
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5,0.5,0.5],
                         std=[0.5,0.5,0.5])
])
```

```python
# SECTION 4 — Prediction Function

def predict(image_path):
    """Run the model on a single image and return class index and confidence."""
    # open image and ensure RGB
    img = Image.open(image_path).convert("RGB")
    # apply preprocessing
    inp = transform(img).unsqueeze(0)  # add batch dim
    with torch.no_grad():
        out = model(inp)
        probs = torch.softmax(out, dim=1)
        conf, idx = torch.max(probs, dim=1)
    # return as python values
    return idx.item(), conf.item()
```
